```python
%load_ext autoreload
%autoreload 2
import matplotlib
matplotlib.use("svg")
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
matplotlib.rc("figure", figsize=(6,4))
import pandas as pd

from tfnn import Neural_Net
from mfcc import create_feature, mel_spectrogram
```

```python
def split_dataset(features, labels, r=0.7):
    p = np.random.permutation(features.shape[0])
    N = int(features.shape[0] * r)
    x_train = features[p][:N]
    y_train = labels[p][:N]
    x_test = features[p][N:]
    y_test = labels[p][N:]
    return x_train, y_train, x_test, y_test
```

```python
dir0 = 'data/female/sober'
dir1 = 'data/female/drunk'
threshold = 0.06
X, y = create_feature(dir0, dir1, 1, threshold, tmax=120, verbose=True)
```

```python
print X.shape
print X[y==0].size
print X[y==1].size
```

```python
plt.plot(range(X.shape[0]), X.mean(axis=1))
plt.show()
```

```python
f, t = np.meshgrid(range(X.shape[1]), range(X.shape[0]))
plt.figure(figsize=[10, 3])
plt.pcolormesh(t, f, X)
plt.axis('tight')
plt.show()
```

```python
df = pd.DataFrame(columns='nframe hsize1 hsize2 lr reg accuracy'.split())
accuracy = {}
idx = 1

for nframe in xrange(5, 10, 2):
    X, y = create_feature(dir0, dir1, nframe, threshold)
    print "#frame: %d, data_size: %d" %(nframe, X.shape[0])
    X_train, y_train, X_test, y_test = split_dataset(X, y)
    for hsize1 in xrange(3, 6):
        for hsize2 in xrange(2, hsize1):
            print "hsize1: %d, hsize2: %d" %(2**hsize1, 2**hsize2)
            for lr in np.logspace(-3, -2, 2):
                for reg in np.logspace(-3, -2, 2):
                    model = Neural_Net([40*nframe, hsize1, hsize2, 1], regularization=reg)
                    sess = model.do_training(X_train, y_train, epochs=30, learning_rate=lr)
                    accuracy = model.do_eval(sess, X_test, y_test)
                    sess.close()
                    df.loc[idx] = (nframe, 2**hsize1, 2**hsize2, lr, reg, accuracy)
                    idx += 1
                    print "lr: %g, reg: %g, acc: %g" %(lr, reg, accuracy)
```

```python
df.groupby(['nframe', 'hsize1', 'hsize2'])['accuracy'].max()
```

```python
def split_dataset2(features, labels, r=0.7):
    X0 = features[labels[:,0]==0]
    X1 = features[labels[:,0]==1]
    N0 = int(X0.shape[0] * r)
    N1 = int(X1.shape[0] * r)
    x_train = np.vstack((X0[:N0], X1[:N1]))
    y_train = np.vstack((np.zeros((N0,1)), np.ones((N1,1))))
    x_test = np.vstack((X0[N0:], X1[N1:]))
    y_test = np.vstack((np.zeros((X0.shape[0]-N0,1)), np.ones((X1.shape[0]-N1,1))))
    return x_train, y_train, x_test, y_test

nframe = 9
X, y = create_feature(dir0, dir1, nframe, threshold)
X_train, y_train, X_test, y_test = split_dataset2(X, y)
print X_train.shape, y_train.shape
print X_test.shape, y_test.shape
```

```python
hsize1 = 32
hsize2 = 8
lr = 0.01
reg = 0.01
model = Neural_Net([40*nframe, hsize1, hsize2, 1], regularization=reg)
sess = model.do_training(X_train, y_train, epochs=50, learning_rate=lr)
model.do_eval(sess, X_test, y_test)
```

```python
from sklearn.metrics import precision_score, recall_score

y_pred, p = model.predict(sess, X_test)
print precision_score(y_test, y_pred)
print recall_score(y_test, y_pred)
```
